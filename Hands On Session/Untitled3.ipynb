```python
x=1

while x<=100:

    if x%15 == 0:
        print("FizzBuzz")
    elif x%5 == 0:
        print("Buzz")
    elif x%3 == 0:
        print("Fizz")
    else:
        print(x)
    
    x+=1
    
```

```python
s = input("Input : ")
n = len(s)

s_back = s[n-1]
a = n-2

while a>=0:
    huruf_a = s[a]
    s_back += huruf_a
    a -= 1

if s == s_back:
    print("Output : Yes")
else:
    print("Output : No")
```
