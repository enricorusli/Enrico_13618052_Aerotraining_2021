```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df['Sex']=df['Sex'].isin(['female']).astype(int)
df.head()
```

```python
df['Cabin']=df['Cabin'].replace(np.nan,'N',regex=True)
df['Cabin']=df['Cabin'].str[0:1]
df['Cabin'].unique()
```

```python
df=df.drop(['PassengerId','Name','Ticket'],axis=1)
df.head()
```

```python
df=pd.get_dummies(df)
df.corr()
```

```python
list(df.columns.values)
```

```python
df=df.drop(['Age','Cabin_A','Cabin_F','Cabin_G','Cabin_T','Embarked_Q'],axis=1)
df.head()
```

```python
x = df[['SibSp','Parch','Pclass','Fare','Sex','Cabin_B','Cabin_C','Cabin_D','Cabin_E','Cabin_N','Embarked_C','Embarked_S']]
y = df[['Survived']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
scaled_data = scaler.transform(x)
scaled_data = pd.DataFrame(scaled_data,columns=x.columns)
scaled_data
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(scaled_data,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(x_train,y_train.values.ravel())
```

```python
from sklearn.preprocessing import StandardScaler

model.score(x_test,y_test)
```

```python
dftest=pd.read_csv("test.csv")
dftest['Sex']=dftest['Sex'].isin(['female']).astype(int)
dftest['Cabin']=dftest['Cabin'].replace(np.nan,'N',regex=True)
dftest['Cabin']=dftest['Cabin'].str[0:1]
xtest=dftest.drop(['PassengerId','Name','Ticket'],axis=1)
xtest=pd.get_dummies(xtest)
xtest = xtest[['SibSp','Parch','Pclass','Fare','Sex','Cabin_B','Cabin_C','Cabin_D','Cabin_E','Cabin_N','Embarked_C','Embarked_S']]
dftest.head()
```

```python
sampledata = xtest.iloc[0,:].tolist()
model.predict([sampledata])
```
