```python
class Isentropic:
    
    def __init__(self,mach,shratio,gconstant):
        self.mach=mach
        self.shratio=shratio
        self.gconstant=gconstant
        
    def p_ratio(self):
        pr = (1+0.5*(self.shratio-1)*self.mach**2)**(-self.shratio/(self.shratio-1))
        print("p/p_0 = " + str(pr))
    
    def rho_ratio(self):
        rr = (1+0.5*(self.shratio-1)*self.mach**2)**(-1/(self.shratio-1))
        print("rho/rho_0 = " + str(rr))
    
    def T_ratio(self):
        tr = (1+0.5*(self.shratio-1)*self.mach**2)**(-1)
        print("T/T_0 = " + str(tr))
        
```
